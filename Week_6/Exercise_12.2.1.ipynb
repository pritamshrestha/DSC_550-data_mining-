```python
import numpy as np
import traceback


def eval_w(w, X, act):
    j = 0
    if act == 'viagra':
        for b in X:
            if b == 1:
                w[j] = w[j] * 2
            j += 1
    elif act == 'nigeria':
        for b in X:
            if b == 1:
                w[j] = w[j] / 2
            j += 1

    return w


def winnow(value_M):
   
    # initialize w
    w = [1, 1, 1, 1, 1, 1]
    iteration = 0
    w_diff = 0
    while w_diff < 2 and iteration < 20:
        wso = sum(w[:5])
        for i in range(len(value_M)):
            wi = sum(w * value_M[i][:-1])
            if wi > 0 and value_M[i][-1] == -1:
                # eval weight
                w = eval_w(w, value_M[i][:-2], act='viagra')
                w[-1] = w[-1] * 2
            elif (wi < 0 and value_M[i][-1] == 1) or wi == 0:
                # eval weight
                w = eval_w(w, value_M[i][:-2], act='nigeria')
                w[-1] = w[-1] / 2
        wsn = sum(w[:5])
        w_diff = wsn - wso
        iteration += 1

    return iteration, w


if __name__ == '__main__':

    # Given data
    given_values = {'a': (1, 1, 0, 1, 1, -1, +1), 'b': (0, 0, 1, 1, 1, -1, -1), 'c': (0, 1, 1, 0, 0, -1, +1),
             'd': (1, 0, 0, 1, 0, -1, -1), 'e': (1, 0, 1, 0, 1, -1, +1), 'f': (1, 0, 1, 1, 0, -1, -1)}

    # Convert to Matrix
    heading = ['a', 'b', 'c', 'd', 'e', 'f']
    value_M = np.array([given_values[i] for i in heading])

    # Display matrix
    print(value_M)

    try:
        # Apply winnow method
        iteration, w = winnow(value_M)
        print('Iteration Count: ', iteration)
        print('Final weight:', w[:5])
    except Exception as exception:
        print('exception')
        traceback.print_exc()
        print('An exception of type {0} occurred. Arguments:\n{1!r}'.format(type(exception).__name__, exception.args))
    finally:
        print("Finally block is executed whether exception is handled or not!!")
```
